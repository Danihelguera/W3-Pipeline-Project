```python
import json
import requests
import pandas as pd
from pandas import json_normalize
import os
from dotenv import load_dotenv
import datetime
import re
import time

load_dotenv()
```

```python
def request_Aemet(enlace):
    aemet = os.getenv("aemet")
    if not aemet:
        raise ValueError("OJO necesitas un token")
    print(f"Estoy haciendo request a {enlace}")
    parametros = {"api_key": aemet}
    res = requests.get(enlace, params=parametros).json()
    return res
```

```python
aemet = os.getenv("aemet")
parametros = {"api_key": aemet}
```

```python
from datetime import date
import time
fecha = date(2010, 1, 1)

url_A = "https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/"
url_C = "T00%3A00%3A00UTC/fechafin/"
url_E = "T23%3A59%3A59UTC/estacion/0201D"

df_clima_BCN = pd.DataFrame()

while fecha <= date(2010,12,31) :
    time.sleep(1)
    url_B = fecha.strftime("%Y-%m-%d")
    res   = requests.get( url_A+url_B+url_C+url_B+url_E , params=parametros).json()
    try :
        datos = requests.get(res['datos']).json()
        datos_df = pd.DataFrame(datos)
        print(f"Leyendo datos meteorologicos en Barcelona el {url_B}")
        df_clima_BCN = pd.concat([df_clima_BCN , datos_df] , ignore_index=True )
    except KeyError:
        print(f"Datos meteorologicos en Barcelona el {url_B} no disponibles")
    except :
        print(f"Datos meteorologicos en Barcelona el {url_B} raros --> pasapalabra")
    fecha = fecha + datetime.timedelta(days=1)
```
