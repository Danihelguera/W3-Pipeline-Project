# Empezamos con el análisis de los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
sns.set(rc = {'figure.figsize':(15,5)})
```

```python
df = pd.read_csv('output/accidents_Bcn_limpio.csv' , encoding='latin-1');
clima_BCN = pd.read_csv('output/clima_BCN.csv')
```

```python
ax=sns.countplot(x='Año',data=df);
ax.set_ylabel('#Accidentes');
ax.set_xlabel('Año');
ax.set_title('Accidentes por año');
```

La cantidad de accidentes se incrementa anualmente.

Habría que revisar si ha aumentado también la población y/o el tráfico.

El año 2020 no es representativo de la tendencia por coincidir con el año del confinamiento

```python
ax=sns.countplot(x='Hora',data=df);
ax.set_ylabel('#Accidentes');
ax.set_xlabel('Hora');
ax.set_title('Accidentes según la hora');
```

Aunque la mayoría de los accidentes son a partir de las 12 del medio dia, se observa claramente el pico de accidentes en al hora punta de las 8 a las 9 de la mañana.

```python
ax=sns.countplot(x='Turno',data=df);
ax.set_ylabel('#Accidentes');
ax.set_xlabel('Momento del día');
ax.set_title('Accidentes según el momento del día');
```

Claramente hay más accidentes por la tarde que por la mañana.

Son datos absolutos, no relativos ya que no se conoce la cantidad de tráfico en cada franja.

```python
ax=sns.countplot(x='Dia_semana',data=df);
ax.set_ylabel('#Accidentes');
ax.set_xlabel('Dia de la semana');
ax.set_title('Accidentes según el día día de la semana');
# order=result['Lunes','Martes','Miercoles','Jueves','Viernes','Sabado','Domingo']
```

## ME FALTA COLOCAR LOS DIAS POR ORDEN

Se ve claramente como la cantidad de accidentes se reduce durante los fines de semana. 

Tampoco se pueden sacar conclusiones sobre una siniestralidad relativa porque se desconoce la densidad de tráfico durante cada día.

```python
df.groupby('Tipo_veh_grup').Año.count().sort_values(ascending=False)
```

```python
ax=sns.countplot(x='Tipo_veh_grup',data=df);
ax.set_ylabel('#Accidentes');
ax.set_xlabel('Tipo de vehículo');
ax.set_title('Accidentes según el momento del día');
```

Proporcionalmente hay muchos accidentes de Motocicletas, sabiendo que no representan ese porcentaje de vehículos en el parque movil. 

Se puede derivar que las Motocicletas tienen una siniestralidad superior al resto de los vehículos.

```python
ax=sns.countplot(x='Color',data=df);
ax.set_ylabel('#Accidentes');
ax.set_xlabel('Color del vehículo');
ax.set_title('Accidentes según el color del vehículo');
```

De este gráfic se pude entender que hay más accidentes de vehículos negros, no obstante, no podemos inferir información sobre la influencia del color en la siniestralidad.

Para saber si el color del vehículo influye en la cantidad de vehículos, se necesitaría conocer la cantidad de vehículos en circulación de cada color y hablar de siniestralidad relativa por cada color.

# Hacemos un mapa de calor para localizar las zonas con más siniestralidad

```python
df_cor = df[['Latitud','Longitud']]
df_cor = df_cor.dropna()

df_cor_fixed = pd.DataFrame(columns=['Latitud', 'Longitud'])

for i in range(len(df_cor)):
    try: 
        df_cor_fixed = df_cor_fixed.append({'Latitud'  : float(df_cor.iloc[[i]]['Latitud']) , 
                                            'Longitud' : float(df_cor.iloc[[i]]['Longitud']) 
                                            }, ignore_index=True)
    except:
        pass
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap

coordinates = [41.390205, 2.154007]

map_acc = folium.Map(location=coordinates, zoom_start = 13)

cor = [[row['Latitud'],row['Longitud']] for index,row in df_cor_fixed.iterrows()]

HeatMap(cor, min_opacity=0.5, radius=14).add_to(map_acc)
map_acc
```

Se identifica claramente las calles principales de Barcelona, Diagonal y Cortes Catalanas como calles con mayor siniestralidad. También son las más concurridas diariamente.
