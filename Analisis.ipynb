# Hacemos un mapa de calor para localizar las zonas con m√°s siniestralidad

```python
df_cor = df[['Latitud','Longitud']]
df_cor = df_cor.dropna()

df_cor_fixed = pd.DataFrame(columns=['Latitud', 'Longitud'])

for i in range(len(df_cor)):
    try: 
        df_cor_fixed = df_cor_fixed.append({'Latitud'  : float(df_cor.iloc[[i]]['Latitud']) , 
                                            'Longitud' : float(df_cor.iloc[[i]]['Longitud']) 
                                            }, ignore_index=True)
    except:
        pass
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap

coordinates = [41.390205, 2.154007]

map_acc = folium.Map(location=coordinates, zoom_start = 13)

cor = [[row['Latitud'],row['Longitud']] for index,row in df_cor_fixed.iterrows()]

HeatMap(cor, min_opacity=0.5, radius=14).add_to(map_acc)
map_acc
```

Se identifica claramente las calles principales de Barcelona, Diagonal y Cortes Catalanas como calles con mayor siniestralidad. Tambi√©n son las m√°s concurridas diariamente.
